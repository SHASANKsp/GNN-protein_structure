```python
import graphein
from graphein.protein.config import ProteinGraphConfig
from graphein.protein.graphs import construct_graph
from graphein.protein.visualisation import plotly_protein_structure_graph
from graphein.protein.edges.distance import (add_peptide_bonds, add_hydrogen_bond_interactions, add_disulfide_interactions, add_ionic_interactions,
                                                add_aromatic_interactions, add_aromatic_sulphur_interactions, add_cation_pi_interactions)
```

```python
ProteinGraphConfig()
```

```python
config = ProteinGraphConfig()
new_edge_funcs = {"edge_construction_functions":[add_peptide_bonds, add_hydrogen_bond_interactions, add_disulfide_interactions, add_cation_pi_interactions],
                 "keep_hets":[False],"insertations":False,"exclude_waters":True,"deprotonate":False,"verbose":True}
config = ProteinGraphConfig(**new_edge_funcs)
```

```python
plotly_protein_structure_graph
```

```python
g =construct_graph(config=config,pdb_code="4ept")
p = plotly_protein_structure_graph(g,colour_edges_by="kind",label_node_ids=False,node_size_multiplier=0.1)
p.show()
```
